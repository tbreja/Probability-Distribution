#       Probability Distribution Introduction

Certain distributions share characteristics, so we separate them into types. The well-defined types of distributions we often deal with have elegant statistics. We distinguish between two big types of distributions based on the type of the possible values for the variable – discrete and continuous.

Discrete distribution have characteristic as following : Have a finite number of outcomes, Use formulas we already talked about, Can add up individual values to determine probability of an interval, Can be expressed with a table, graph or a piece-wise function, Expected Values might be unattainable, Graph consists of bars lined up one after the other.
Discrete Distribution consists of these types : Uniform Distribution, Bernoulli, Binomial Distribution, and Poisson Distribution.

On the other hand, Continuous distribution characteristics as following : Have infinitely many consecutive possible values, Use new formulas for attaining the probability of specific values and intervals, Cannot add up the individual values that make up an interval because there are infinitely many of them, Can be expressed with a graph or a continuous function, Graph consists of a smooth curve.
Continuous Distribution consists of these types : Normal Distribution, Student's T Distribution, Chi-Squared Distribution, Exponential Distribution and Logistic Distribution.

below, we have some example of fistribution types from FIFA 19 dataset. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('FIFA19.csv', delimiter=',')
df.head()
```

## Overall Value Distribution

```python
 # To begin with examine the overall column
 bin_list = list(range(46,94,1))
 ovrl = plt.hist(df['Overall'],bins= bin_list )
 ovrl = plt.title('Overall Value Normal Distibution')
 
```

the graph is bell-shaped with thin tails, from the grapgh we can conclude that : 
1. overall value has normal distribution. 
2. Now we can notice the thin tails suggest a smaller number of outliers. 
3. This reflects real life quite accurately since very few professional players are exceptionally good or bad at every single aspect of the sport besides even the least skilled professional soccer players are far superior to the average person. 
4. That explains why the lowest overall values start from around 50 rather than 0 the stats should reflect. 
5. Thus we can safely consider the game balanced and acceptable for competitive play.

if we take some of sample from overall value (which is normal distribution). we can have a student's T distribution. for example we would like to take first 100 sample. 

```python
# Overal value first 100 sample
bin_list = list(range(53,89,3))
ovrl = plt.hist(df['Overall'].head(30), bins=bin_list)
ovrl = plt.title('Overall Value Students T Distribution')
```

We will see that the graph slightly resembles a normal distribution. However we will also notice the fatter tails since the number of observations is limited. We can safely consider this sample follows a student's t distribution recall that the Student's t distribution is also symmetric.

# Real Face values Distribution

```python
real_face = df['Real Face']
real_face.dropna(inplace=True)
plt.title('Real Face values Distribution')
plt.hist(real_face)
```

The graph consist from 2 bars, it means that Real Face value only have 2 outcomes which is 'Yes' or 'No'. from the characteristic we can safely conclude that Real Face variable is Bernoulli Distribution.

if we plot the histogram of Preffered Foot value we should get Bernoulli Dsitribution too, considering the outcomes has only 2 types, which is 'right' or 'left'.
